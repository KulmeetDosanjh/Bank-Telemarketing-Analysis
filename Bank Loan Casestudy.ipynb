```python
temporary.index
```

```python
Corr_Table=pd.DataFrame()
lst_var1=[]
lst_var2=[]
lst_var3=[]
for i in temporary.index:
    for j in temporary.columns:
        if i==j or temporary.loc[i,j]==np.NaN:
            
            continue
        else:
            lst_var1.append(i)
            lst_var2.append(j)
            lst_var3.append(temporary.loc[i,j])

Corr_Table["Variable1"]=lst_var1
Corr_Table["Variable2"]=lst_var2  
Corr_Table["corr_value"]=lst_var3  
```

```python
Corr_Table=Corr_Table[~(Corr_Table["corr_value"].isnull())]
```
